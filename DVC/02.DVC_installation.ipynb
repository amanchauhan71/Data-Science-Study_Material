# 4. DVC
Data Version Control, or DVC, is a data and ML experiment management tool that takes advantage of the existing engineering toolset that you're already familiar with (Git, CI/CD, etc.).

## 4.1 Installation
DVC can be used as a Python library, simply install it with a package manager like pip or conda, and as a Python project requirement if needed. The Python API module is dvc.api.



### 4.1.1 Install on Linux

To use DVC as a [Python library](https://dvc.org/doc/api-reference), please install with pip or with conda.

**Install with pip**

We strongly recommend creating a virtual environment or using [pipx](https://packaging.python.org/guides/installing-stand-alone-command-line-tools/) (on Python 3.6+) to encapsulate your local environment.

```shell
$ pip install dvc
```

**Install with conda**

Requires [**Miniconda**](https://docs.conda.io/en/latest/miniconda.html) or **Anaconda Distribution**



```bash
$ conda install -c conda-forge mamba  # installs much faster than conda
$ mamba install -c conda-forge dvc
```

**Install with snap**

Requires 
[**snapd**](https://snapcraft.io/docs/installing-snapd)


```bash
$ snap install --classic dvc
```

**Install from repository**

On Debian/Ubuntu

```bash
sudo wget \
  https://dvc.org/deb/dvc.list \
  -O /etc/apt/sources.list.d/dvc.list
wget -qO - https://dvc.org/deb/iterative.asc | sudo apt-key add -
sudo apt update
sudo apt install dvc
```




On Fedora/CentOS

```bash
sudo wget \
 https://dvc.org/rpm/dvc.repo \
 -O /etc/yum.repos.d/dvc.repo
sudo rpm --import https://dvc.org/rpm/iterative.asc
sudo yum update
sudo yum install dvc
```

**Install from package**

Get the binary package from the big "Download" button on the [home page](https://dvc.org/), or from the [release page](https://github.com/iterative/dvc/releases/) on GitHub. Then install it with the following command.

On Debian/Ubuntu

```bash
sudo apt install ./dvc_0.62.1_amd64.deb
```

On Fedora/CentOS

```bash
sudo yum install dvc-0.62.1-1.x86_64.rpm
```

### 4.1.2 Installation on Windows

To use DVC as a [Python library](https://dvc.org/doc/api-reference), please install with pip or with conda.

**Install with choco**

The easiest way to install from command line for most cases is to install [Chocolatey](https://chocolatey.org/) on your machine, and use the choco command:

```bash
$ choco install dvc
```

**Install with conda**

Requires [Miniconda](https://docs.conda.io/en/latest/miniconda.html) or Anaconda Navigator.

You can use conda from [Anaconda Prompt](https://docs.anaconda.com/anaconda/user-guide/getting-started/#open-prompt-win), a POSIX-like command line terminal in Windows.

```bash
$ conda install -c conda-forge mamba # installs much faster than conda
$ mamba install -c conda-forge dvc
```

**Install with pip**

From Command Prompt or other [recommended consoles](https://dvc.org/doc/user-guide/running-dvc-on-windows):

```bash
$ pip install dvc
```

**Windows installer**

An easy way is to use the self-contained, executable installer (binary), which is available from the big "Download" button on the [home page](https://dvc.org/). You can also get it from the [release page](https://github.com/iterative/dvc/releases/) on GitHub.

You'll need to download and run the installer again each time you want to update DVC. You may use Windows Uninstaller to [remove the program](https://dvc.org/doc/install/windows).

### 4.1.3 Installation on macOS

To use DVC as a [Python library](https://dvc.org/doc/api-reference), please install with pip or with conda.

**Install with brew**

Recommended. Requires [Homebrew](https://brew.sh/)

```bash
$ brew install dvc
```

**Install from package**

Get the PKG (binary) from the big "Download" button on the [home page](https://dvc.org/), or from the [release page](https://github.com/iterative/dvc/releases/) on GitHub.

**Install with pip**

We strongly recommend creating a virtual environment or using [pipx](https://packaging.python.org/guides/installing-stand-alone-command-line-tools/) (on Python 3.6+) to encapsulate your local environment.

```bash
$ pip install dvc
```

**Install with conda**

Requires [Miniconda](https://docs.conda.io/en/latest/miniconda.html) or Anaconda Distribution

```bash
$ conda install -c conda-forge mamba # installs much faster than conda
$ mamba install -c conda-forge dvc
```
